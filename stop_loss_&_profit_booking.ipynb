```python
# ======================== Seperate Loss and Profit ===========================

import numpy as np
import pandas as pd
import scipy.stats as stats

# 1) PREPARE & FILTER OPTIONS
opts['date']   = pd.to_datetime(opts['date'])
opts['exdate'] = pd.to_datetime(opts['exdate'])
opts['mid']    = (opts['adj_best_bid'] + opts['adj_best_offer']) / 2

contracts = (
    test_4[['ticker','expirationDate','adjStrikePrice']]
        .drop_duplicates()
        .rename(columns={
            'expirationDate':'exdate',
            'adjStrikePrice':'adj_strike_price'
        })
)
opts = opts.merge(contracts, on=['ticker','exdate','adj_strike_price'], how='inner')

# 2) ANNOTATE EACH QUOTE WITH STRADDLE ID
test_4 = test_4.reset_index().rename(columns={'index':'sid'})
merged = pd.merge(
    test_4[['sid','ticker','tradingDate','expirationDate','adjStrikePrice','weightCall','weightPut','entry']],
    opts[['ticker','date','exdate','adj_strike_price','cp_flag','mid']],
    left_on=['ticker','expirationDate','adjStrikePrice'],
    right_on=['ticker','exdate','adj_strike_price'],
    how='inner'
)
merged = merged[
    (merged['date'] >= merged['tradingDate']) &
    (merged['date'] <= merged['expirationDate'])
].reset_index(drop=True)

# 3) PIVOT INTO CALL/PUT MIDS AND COMPUTE COMBINED
merged['cp_flag'] = merged['cp_flag'].str.upper().map({'C':'C','P':'P'})
pivoted = (
    merged
      .pivot(index=['sid','date'], columns='cp_flag', values='mid')
      .reset_index()
)
pivoted.columns.name = None
pivoted = pivoted.dropna(subset=['C','P'])
pivoted = pivoted.merge(
    test_4[['sid','weightCall','weightPut','entry','tradingDate']],
    on='sid', how='left'
)
pivoted['combined'] = pivoted['weightCall']*pivoted['C'] + pivoted['weightPut']*pivoted['P']

# capture entry‐leg mids
entry_mids = pivoted[pivoted['date']==pivoted['tradingDate']][['sid','C','P']] \
                  .rename(columns={'C':'C_entry','P':'P_entry'})

pivoted = pivoted.merge(entry_mids, on='sid', how='left')

# 4) FIND FIRST DATES FOR PROFIT‐BOOKING & LEG STOP‐LOSS
profit_dates   = pivoted[pivoted['combined'] <= pivoted['entry']*0.01] \
                    .groupby('sid')['date'].min().rename('profit_date')
c_stop_dates   = pivoted[pivoted['C'] >= pivoted['C_entry']*1.75] \
                    .groupby('sid')['date'].min().rename('c_stop_date')
p_stop_dates   = pivoted[pivoted['P'] >= pivoted['P_entry']*1.75] \
                    .groupby('sid')['date'].min().rename('p_stop_date')
expiry_dates   = pivoted.groupby('sid')['date'].max().rename('expiry_date')

events = pd.concat(
    [profit_dates, c_stop_dates, p_stop_dates, expiry_dates],
    axis=1
).reset_index()

def choose_exits(row):
    pdt, csd, psd, xpd = (
        row['profit_date'],
        row['c_stop_date'],
        row['p_stop_date'],
        row['expiry_date']
    )
    # 1) Profit booking if first
    if pd.notna(pdt) and pdt <= (csd if pd.notna(csd) else xpd) and pdt <= (psd if pd.notna(psd) else xpd):
        return pd.Series({'call_exit_date':pdt,'put_exit_date':pdt,'exit_type':'profit'})
    # 2) Else each leg stops on its own stop date or expiry
    ced = csd if pd.notna(csd) and csd <= xpd else xpd
    ped = psd if pd.notna(psd) and psd <= xpd else xpd
    etype = 'expiry' if (ced==xpd and ped==xpd) else 'stop_loss'
    return pd.Series({'call_exit_date':ced,'put_exit_date':ped,'exit_type':etype})

events[['call_exit_date','put_exit_date','exit_type']] = events.apply(choose_exits, axis=1)

# 5) LOOK UP EXIT‐LEG MIDS
ce = (
    events[['sid','call_exit_date']]
      .merge(pivoted[['sid','date','C']],
             left_on=['sid','call_exit_date'],
             right_on=['sid','date'],
             how='left')
      .loc[:,['sid','C']]
      .rename(columns={'C':'C_exit'})
      .set_index('sid')
)
pe = (
    events[['sid','put_exit_date']]
      .merge(pivoted[['sid','date','P']],
             left_on=['sid','put_exit_date'],
             right_on=['sid','date'],
             how='left')
      .loc[:,['sid','P']]
      .rename(columns={'P':'P_exit'})
      .set_index('sid')
)
ev = events.set_index('sid').join(ce).join(pe)

# 6) BUILD RAW & JOIN EVENTS + ENTRY MIDS + EXIT MIDS
raw = (
    test_4.set_index('sid')
          # bring in exit dates, exit_type, AND the two exit mids
          .join(ev[['call_exit_date','put_exit_date','exit_type','C_exit','P_exit']])
          # bring in the entry‐leg mids
          .join(entry_mids.set_index('sid'))
)

# 7) MERGE STOCK EXPIRY PRICE & COMPUTE INTRINSIC PAYOFFS
stk2 = stock_data.copy(); stk2['date']=pd.to_datetime(stk2['date'])
exp_price = stk2.rename(columns={'date':'expirationDate','adj_prc':'prc_stk_expire'})[
    ['ticker','expirationDate','prc_stk_expire']
]
raw = raw.reset_index().merge(exp_price,
       on=['ticker','expirationDate'],how='left').set_index('sid')
raw['call_payoff'] = np.maximum(raw['prc_stk_expire'] - raw['adjStrikePrice'],0)
raw['put_payoff']  = np.maximum(raw['adjStrikePrice'] - raw['prc_stk_expire'],0)

# 8) COMPUTE LEG‐BY‐LEG PROFIT USING STOP VS EXPIRY PER LEG
raw['profit_call'] = np.where(
    raw['call_exit_date'] < raw['expirationDate'],
    # call stopped early: use mid‐difference
    raw['weightCall'] * (raw['C_entry'] - raw['C_exit']),
    # call expired: use intrinsic payoff
    raw['weightCall'] * (raw['C_entry'] - raw['call_payoff'])
)

raw['profit_put'] = np.where(
    raw['put_exit_date'] < raw['expirationDate'],
    # put stopped early: use mid‐difference
    raw['weightPut'] * (raw['P_entry'] - raw['P_exit']),
    # put expired: use intrinsic payoff
    raw['weightPut'] * (raw['P_entry'] - raw['put_payoff'])
)

raw['total_profit'] = raw['profit_call'] + raw['profit_put']
raw['final_return'] = raw['total_profit'] / raw['entry']

# 9) (Optional) stock prices at entry, exit, next‐day…

# 10) ASSEMBLE FINAL TABLE
stop_results = raw.reset_index()[[
    'sid','ticker','tradingDate','adjStrikePrice','weightCall','weightPut','entry',
    'expirationDate','exit_type',
    'call_exit_date','C_entry','C_exit','call_payoff','profit_call',
    'put_exit_date','P_entry','P_exit','put_payoff','profit_put',
    'total_profit','final_return'
]]
print(stop_results.head())
```

```python
from IPython.display import display, Math

# 1) Entry
display(Math(r'''
\mathrm{entry}
\;=\;
w_C\;\frac{\mathrm{bid}_{\mathrm{call}}+\mathrm{ask}_{\mathrm{call}}}{2}
\;+\;
w_P\;\frac{\mathrm{bid}_{\mathrm{put}}+\mathrm{ask}_{\mathrm{put}}}{2}
'''))

# 2) Early Exit (Stop‐loss)
display(Math(r'''
\mathrm{combined}_t
\;=\;
w_C\,\mathrm{mid}_{\mathrm{call}}(t)
\;+\;
w_P\,\mathrm{mid}_{\mathrm{put}}(t)
'''))
display(Math(r'''
\text{if }\mathrm{combined}_t \ge 1.5\times\mathrm{entry}\text{ at }t_s:
\quad
\begin{cases}
\mathrm{exit\_cost}=\mathrm{combined}_{t_s},\\
\mathrm{stop\_profit}=\mathrm{entry}-\mathrm{exit\_cost},\\
R_{\mathrm{stop}}
=\displaystyle\frac{\mathrm{stop\_profit}}{\mathrm{entry}}.
\end{cases}
'''))

# 3) Hold‐to‐Expiry (No Stop)
display(Math(r'''
\mathrm{call\_payoff}=\max(S_T-K,0),
\quad
\mathrm{put\_payoff}=\max(K-S_T,0),
\quad
\mathrm{payoff}=w_C\,\mathrm{call\_payoff}+w_P\,\mathrm{put\_payoff}
'''))
display(Math(r'''
\mathrm{hold\_profit}=\mathrm{entry}-\mathrm{payoff},
\quad
R_{\mathrm{hold}}
=\displaystyle\frac{\mathrm{hold\_profit}}{\mathrm{entry}}.
'''))

# 4) Final Return
display(Math(r'''
R = 
\begin{cases}
R_{\mathrm{stop}}, & \text{if stop‐loss triggered},\\[6pt]
R_{\mathrm{hold}}, & \text{otherwise (expiration)}.
\end{cases}
'''))
```
